```python
# Setup (Paste & Run This First!)
import subprocess, sys, os
from getpass import getpass

# Install
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "langchain openai langchain-openai"])

# API Key
api_key = getpass("Paste OpenAI API Key: ")
os.environ["OPENAI_API_KEY"] = api_key

# Import
from langchain_openai import ChatOpenAI
print("✅ Setup Complete!")

# =================== HELLO WORLD ===================
llm = ChatOpenAI(temperature=0.7, model="gpt-3.5-turbo")
response = llm.invoke("Jelaskan LangChain dalam 1 kalimat!")
print("\n🤖 Response:", response.content)
```
